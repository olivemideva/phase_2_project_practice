### 3. Checking and dealing with outliers

Addressing outliers and improving data skewness before building linear and multilinear regression models is critical for enhancing model accuracy and reliability. Removing outliers reduces the impact of extreme data points on model fitting, leading to more accurate predictions and parameter estimates. Handling skewness ensures that the data distribution aligns better with regression assumptions, such as normality of residuals, improving the validity of statistical inferences and model interpretation. By preprocessing for outliers and skewness, regression models become more stable, performant, and capable of capturing meaningful relationships between predictors and the target variable. Overall, these steps optimize model performance and enhance the quality of insights derived from regression analysis.

```python
sns.catplot(x='price', data=df, kind='box', height=3, aspect=3).set(title="With Outliers")
```

```python
Q75 = np.percentile(df['price'],75)
Q25 = np.percentile(df['price'],25)
IQR = Q75-Q25
cutoff = IQR * 1.5
upper = Q75 + cutoff
lower = 1

df = df[(df['price']<upper)]
df.head()
```

### MODELLING

***1. Linear Regression model***

We will pick **`sqft_living` - Square footage of living space in the home** to be used to create our linear regression model because it has the most correlation with the price and it has the most linear scatter plor hence a good candidate.

```python
# Prepare y and X for modeling
X_linear = df[['sqft_living']]
y = df['price']
# Adding a constant column to the features matrix for both models
X_linear_with_const = np.column_stack([np.ones(len(X_linear)), X_linear])
```

```python
linear_model = LinearRegression()
linear_model.fit(X_linear_with_const, y)
```

***2. Multilinear Regression model***

```python
# Prepare y and X for modeling
X_multilinear = df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot','condition']]
y = df['price']
```

```python
# One-Hot encoding
X_multilinear = pd.get_dummies(X_multilinear, columns=["condition"], drop_first=True, dtype=int) 
```

```python
# Adding a constant column to the features matrix for both models
X_multilinear_with_const = np.column_stack([np.ones(len(X_multilinear)), X_multilinear])
```

```python
multilinear_model = LinearRegression()
multilinear_model.fit(X_multilinear_with_const, y)
```

### Evaluation of the models

***Predicted values***

```python
# Make predictions with linear_model
predictions_linear = linear_model.predict(X_linear_with_const)
# Make predictions with multilinear_model
predictions_multilinear = multilinear_model.predict(X_multilinear_with_const)
```

***Linear Model Summary***

```python
# Calculating R-squared
r2_linear = r2_score(y, predictions_linear)

# Model summaries
print("Linear Model Summary:")
print("Intercept:", linear_model.intercept_)
print("Coefficient:", linear_model.coef_[1])
print("R-squared:", r2_linear)
```

***Multilinear Model Summary***

```python
# Calculating R-squared
r2_multilinear = r2_score(y, predictions_multilinear)

# Model summaries
print("\nMultilinear Model Summary:")
print("Intercept:", multilinear_model.intercept_)
print("Coefficients:", multilinear_model.coef_[1:])
print("R-squared:", r2_multilinear)
```
